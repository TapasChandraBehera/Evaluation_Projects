Now we merge the 2 datasets into the one dataset.

```python
# Merge datasets based on Country Code
merged_df = pd.merge(zomato_df, country_code_df, on='Country Code', how='left')
```

Since we are getting there is some missing values are present inside the merged_df dataset, first we have to handle this.

```python
merged_df['Cuisines'].fillna('Not Specified', inplace=True)
```

Handling categorical columns and numerical columns.

```python
import warnings
from sklearn.preprocessing import OneHotEncoder, StandardScaler
import pandas as pd

# Define categorical columns and numerical columns based on your dataset
categorical_columns = ['City', 'Cuisines', 'Has Table booking', 'Has Online delivery', 'Is delivering now', 'Switch to order menu', 'Currency', 'Rating color', 'Rating text', 'Country']
numerical_columns = ['Average Cost for two', 'Price range', 'Aggregate rating', 'Votes', 'Longitude', 'Latitude']

# Suppress FutureWarnings temporarily
with warnings.catch_warnings():
    warnings.filterwarnings("ignore", category=FutureWarning)
    
    # Perform one-hot encoding for categorical columns
    encoder = OneHotEncoder(sparse_output=False)
    encoded_categorical = encoder.fit_transform(merged_df[categorical_columns])
    encoded_categorical_df = pd.DataFrame(encoded_categorical, columns=encoder.get_feature_names_out(categorical_columns))

# Scale numerical columns
scaler = StandardScaler()
scaled_numerical = scaler.fit_transform(merged_df[numerical_columns])
scaled_numerical_df = pd.DataFrame(scaled_numerical, columns=numerical_columns)

# Concatenate encoded categorical and scaled numerical dataframes
preprocessed_df = pd.concat([encoded_categorical_df, scaled_numerical_df], axis=1)
```

Exploratory Data Analysis(EDA).

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Compute summary statistics for numerical features
numerical_summary = preprocessed_df[numerical_columns].describe()

# Visualize distributions of numerical features using histograms
plt.figure(figsize=(12, 8))
for i, column in enumerate(numerical_columns):
    plt.subplot(3, 2, i+1)
    sns.histplot(preprocessed_df[column], kde=True)
    plt.title(f'{column} Distribution')
    plt.xlabel(column)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# Visualize box plots for numerical features to identify outliers
plt.figure(figsize=(12, 8))
sns.boxplot(data=preprocessed_df[numerical_columns])
plt.title('Box Plot of Numerical Features')
plt.xlabel('Features')
plt.ylabel('Values')
plt.xticks(rotation=45)
plt.show()
```

Model Building.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.metrics import mean_absolute_error, accuracy_score
from sklearn.preprocessing import LabelEncoder
```

```python
# Define target variables
target_variable_regression = 'Average Cost for two'
target_variable_classification = 'Price range'
```

```python
# Separate features and target variables
X = preprocessed_df.drop([target_variable_regression, target_variable_classification], axis=1)
y_regression = preprocessed_df[target_variable_regression]
y_classification = preprocessed_df[target_variable_classification]
```

```python
# Split data into training and testing sets for regression
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X, y_regression, test_size=0.2, random_state=42)
```

```python
# Split data into training and testing sets for classification
X_train_cls, X_test_cls, y_train_cls, y_test_cls = train_test_split(X, y_classification, test_size=0.2, random_state=42)
```

```python
# Choose machine learning algorithms
regression_model = LinearRegression()  # Example: Linear Regression for regression task
classification_model = LogisticRegression(max_iter=1000)  # Example: Logistic Regression for classification task
```

```python
# Train regression model
regression_model.fit(X_train_reg, y_train_reg)

# Predictions for regression
y_pred_reg = regression_model.predict(X_test_reg)

# Evaluate regression model
mae_regression = mean_absolute_error(y_test_reg, y_pred_reg)
```

```python
# Convert 'Price range' to categorical labels
label_encoder = LabelEncoder()
y_train_classification_encoded = label_encoder.fit_transform(y_train_cls)
y_test_classification_encoded = label_encoder.transform(y_test_cls)
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score

# Convert 'Price range' to categorical labels
label_encoder = LabelEncoder()
y_train_classification_encoded = label_encoder.fit_transform(y_train_cls)
y_test_classification_encoded = label_encoder.transform(y_test_cls)

# Choose machine learning algorithms
classification_model = LogisticRegression(max_iter=1000)  # Example: Logistic Regression for classification task

# Train classification model
classification_model.fit(X_train_cls, y_train_classification_encoded)

# Predictions for classification
y_pred_cls_encoded = classification_model.predict(X_test_cls)

# Decode predicted labels back to original format
y_pred_cls = label_encoder.inverse_transform(y_pred_cls_encoded)

# Evaluate classification model
accuracy_classification = accuracy_score(y_test_classification_encoded, y_pred_cls_encoded)
```

```python
print(f"Mean Absolute Error (Regression): {mae_regression}")
print(f"Accuracy (Classification): {accuracy_classification}")
```

It seems like there might be an issue with the model performance metrics.

```python
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score

# Calculate additional classification metrics
precision = precision_score(y_test_classification_encoded, y_pred_cls_encoded, average='weighted')
recall = recall_score(y_test_classification_encoded, y_pred_cls_encoded, average='weighted')
f1 = f1_score(y_test_classification_encoded, y_pred_cls_encoded, average='weighted')

# Print classification metrics
print(f"Accuracy (Classification): {accuracy_classification}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1-score: {f1}")

# Print confusion matrix
conf_matrix = confusion_matrix(y_test_classification_encoded, y_pred_cls_encoded)
print("Confusion Matrix:")
print(conf_matrix)
```
