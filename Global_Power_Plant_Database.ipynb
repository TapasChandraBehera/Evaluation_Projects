Since our dataset having many missing values and also having many categorical columns let first handle this all.

```python
# Fill missing values in numerical columns with mean
numerical_cols = ['capacity_mw', 'latitude', 'longitude', 'commissioning_year', 'generation_gwh_2013', 'generation_gwh_2014', 'generation_gwh_2015', 'generation_gwh_2016', 'generation_gwh_2017', 'generation_gwh_2018', 'generation_gwh_2019']
combined_data[numerical_cols] = combined_data[numerical_cols].fillna(combined_data[numerical_cols].mean(numeric_only=True))
```

```python
# Fill missing values in categorical columns with 'Unknown'
categorical_cols = ['primary_fuel', 'other_fuel1', 'other_fuel2', 'other_fuel3', 'owner', 'source', 'geolocation_source', 'generation_data_source']
combined_data[categorical_cols] = combined_data[categorical_cols].fillna('Unknown')
```
