```python
#Scale numerical features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Train and test the models using scaled data
logreg_model.fit(X_train_scaled, y_train)
rf_model.fit(X_train_scaled, y_train)
```

```python
# Make predictions on scaled test data
y_pred_logreg_scaled = logreg_model.predict(X_test_scaled)
y_pred_rf_scaled = rf_model.predict(X_test_scaled)
```

```python
# Evaluate model performance on scaled data
accuracy_logreg_scaled = accuracy_score(y_test, y_pred_logreg_scaled)
accuracy_rf_scaled = accuracy_score(y_test, y_pred_rf_scaled)

print(f'Scaled Logistic Regression Accuracy: {accuracy_logreg_scaled:.2f}')
print(f'Scaled Random Forest Accuracy: {accuracy_rf_scaled:.2f}')
```

Checking for underfitting/overfitting.

```python
# Check for underfitting/overfitting with cross-validation
cv_scores_logreg = cross_val_score(logreg_model, X_train_scaled, y_train, cv=5)
cv_scores_rf = cross_val_score(rf_model, X_train_scaled, y_train, cv=5)

print(f'Scaled Logistic Regression CV Scores: {cv_scores_logreg}')
print(f'Scaled Random Forest CV Scores: {cv_scores_rf}')
```

Hyper Parameter tuning.

```python
from sklearn.model_selection import GridSearchCV

# Hyperparameter tuning for Logistic Regression
param_grid_logreg = {'C': [0.01, 0.1, 1, 10, 100]}
grid_logreg = GridSearchCV(logreg_model, param_grid=param_grid_logreg, cv=5)
grid_logreg.fit(X_train_scaled, y_train)

# Hyperparameter tuning for Random Forest
param_grid_rf = {'n_estimators': [100, 200, 300], 'max_depth': [5, 10, 15]}
grid_rf = GridSearchCV(rf_model, param_grid=param_grid_rf, cv=5)
grid_rf.fit(X_train_scaled, y_train)

# Get best parameters
best_params_logreg = grid_logreg.best_params_
best_params_rf = grid_rf.best_params_

print(f'Best Logistic Regression Parameters: {best_params_logreg}')
print(f'Best Random Forest Parameters: {best_params_rf}')
```
